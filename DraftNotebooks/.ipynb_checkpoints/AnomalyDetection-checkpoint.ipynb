###Obtain spread/headway between successive trips at specified distances along pattern 
--capture trip, measure, time-period, headway in seconds, 

```python
def lin_modeler(trip):
    clf = linear_model.LinearRegression()
    x = df_1_S00[df_1_S00['TRIP_TAG'] == trip].PATTERN_MEASURE.values
    y = df_1_S00[df_1_S00['TRIP_TAG'] == trip].SECONDSAFTERMIDNIGHT.values
    return list([x, y, clf.fit(x[:,np.newaxis],y)])#,columns=['X','Y','clf'])
```

```python
def predict_arrival_at_measure(row,measure):
    return row[2].predict(measure)
```

```python
def plot_all_trips(trips):
    plt.figure(figsize=(10,8))
    for index, vals in trips.iteritems():                          
       plt.scatter(vals[0], vals[1])
       plt.plot(vals[0], vals[2].predict(vals[0][:,np.newaxis]))
```

```python
def build_hist(data):
    sns.set_palette("hls", 1)
    plt.ylim(0, 1)
    sns.rugplot(data[8])
```

```python
def build_hist2(data):
    plt.hist(data.values, alpha=.3)
    sns.rugplot(data.values);
```

```python
df_1_S00 = df_1_S00.reset_index().sort('TRIPFROMTIME')
ls = df_1_S00.TRIP_TAG.drop_duplicates('TRIP_TAG').apply(lin_modeler)
```

```python
#Make plots
plot_all_trips(ls)
```

```python
#Get list of headways
build_hist2( ls.apply(predict_arrival_at_measure,measure=1000).diff().dropna() )
```
