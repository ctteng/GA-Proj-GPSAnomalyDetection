```python
 def return_group(df):
    df.sort(['VEHICLE_TAG','TRIP_TAG','REPORT_TIME'],inplace=True)
    #df.groupby(['VEHICLE_TAG','TRIP_TAG','REPORT_TIME'], group_keys=False).apply(process_trips)
    return df
```

```python
def process_trips(dfA):
    #reminder - need way to catch first entry into each group
    cur_position = dfA.index[0]
    prev_position = cur_position - 1
    
    #Trip
    if(cur_position < 1):  
        return dfA
    else:
#        #how can i grab parent dataframe?
        #dfA['DELTA_MEASURE'] =  dfA['PATTERN_MEASURE'] - df_trip_traces.ix[prev_position]['PATTERN_MEASURE']
        dfA['DELTA_MEASURE'] =  dfA['PATTERN_MEASURE'] - df_trip_traces.ix[prev_position]['PATTERN_MEASURE']
        return dfA
```

```python
df_trip_traces = pd.read_csv("C:\\Users\\cteng.G2WZKV1\\Documents\\GitHub\\GA-Proj-GPSAnomalyDetection\\pattern2420150205_vehicle.txt")
df_trip_traces['REPORT_TIME'] = pd.to_datetime(pd.Series(df_trip_traces['REPORT_TIME']))
df_trip_traces['DELTA_MEASURE'] = 0
df_trip_traces['DELTA_TIME'] = 0
df_trip_traces.sort(['VEHICLE_TAG','TRIP_TAG','REPORT_TIME'],inplace=True)
df_trip_traces.reset_index(inplace=True)
```

```python
df_grouped_by = df_trip_traces.groupby(['VEHICLE_TAG','TRIP_TAG'], group_keys=False).apply(return_group).groupby(['VEHICLE_TAG','TRIP_TAG','REPORT_TIME'], group_keys=False).apply(process_trips)
```
